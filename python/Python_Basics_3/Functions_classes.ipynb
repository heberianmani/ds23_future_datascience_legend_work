```python
#Return unique sorted items only #Remove duplicates, filter even and sort 
def returnSortedEven(lst):
    maxNum = max(lst)
    return_set = set(range(0,maxNum+1,2))
    return sorted(list(return_set.intersection(lst)))
```

```python
lst =[1,2,6,4,5,4,5,7,8,2,4,5,9]
returnSortedEven(lst)
```

```python
#Function overloading is to have dynamic parameters and handling
#Ability to degine multiple functions with same name but different parameters or arugment types in a PLanguage
def add(datatype, *args, **kwargs):
    if datatype == 'int':
        ans= 0
        for i in args: 
            ans = ans+i
    
    if datatype == "str":
        ans = ''
        
        for i in args: 
            ans = ans+' '+i
    print(ans)
```
