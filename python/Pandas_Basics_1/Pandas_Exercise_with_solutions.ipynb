## Exercise 2: Write a Pandas program to select the rows where number of attempts in the examination is less than 2 and score greater than 15.

```python
import numpy as np
import pandas as pd
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a','b','c','d','e','f','g','h','i','j']
df = pd.DataFrame(exam_data, index = labels)
df1  = df.loc[(df['attempts'] < 2) & (df['score'] > 15)]
print("Number of attempts in the examination is less than 2 and score greater than 15 :")
print(df1)
print("\n Another solution")
print(df1.query('attempts<2' and 'score>15'))
print("\n Another solution")
print(df.loc[(df['attempts']<2) & (df['score']>15)])
```

## Exerise 3: Change the value in row 'd' to 11.5

```python
import numpy as np
import pandas as pd
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a','b','c','d','e','f','g','h','i','j']
df = pd.DataFrame(exam_data, index = labels)
print("Original data frame : ")
print(df)
print("\nChange the score in row 'd' to 11.5 using loc : ")
df.loc['d', 'score'] = 11.5
print(df)
print("\nChange the score in row 'd' to 11.5 using iloc : ")
df.iloc[3,1] = 11.5
print(df)
```

## Exercise 4: Write a Pandas program to select examination attempts(col1) by the students with 4 attempts.

```python
import numpy as np
import pandas as pd
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d)
print("Original DataFrame")
print(df)
print("\nRows for column1 value == 4")
print(df[df['col1'] == 4])
```

## Exercise 5: Drop a list of specified rows(2nd and 4th) from a DataFrame

```python
import numpy as np
import pandas as pd
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d, index = ['a','b','c','d','e'])
print("Original DataFrame")
print(df)
result = df.drop(df.index[[2,4]])
print('Drop a list of specified rows (2nd and 4th) : ')
print(result)
```

## Exercise 6: Check whether a given column is present in a DataFrame or not

```python
import numpy as np
import pandas as pd
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d)
print("Original DataFrame")
print(df)

if 'col1' in df.columns:
  print('Col1 is present in DataFrame')
else:
  print('Col1 is not present in DataFrame')

if 'col4' in df.columns:
  print('Col4 is present in DataFrame')
else:
  print('Col4 is not present in DataFrame')
```

## Exercise 8: Write a Pandas program to select all columns, except one given column in a DataFrame.

```python
import numpy as np
import pandas as pd
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d)
print("Original DataFrame")
print(df)
print("\n All Columns except 'col3' ")
print(df[['col1', 'col2']])
print("\n Another solution")
print(df.loc[:,df.columns!='col3'])
```

## Exercise 9: Write a Pandas program to remove last n rows of a given DataFrame.

```python
import numpy as np
import pandas as pd
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d)
print("Original DataFrame")
print(df)
print("\n Remove last n rows in a DataFrame")
print(df.head(-2))
print("\n Another solution")
print(df.iloc[:-2])
```
