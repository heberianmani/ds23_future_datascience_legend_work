## Numpy Array Slicing

```python
import numpy as np

a = np.array([1,2,3,4,5,6,7,8])
```

```python
#Slicing syntax = Arrname[start:stop:step]
print(a[0:5])
```

```python
print(a[:5])
```

```python
print(a[3:])
```

```python
print(a[2:5])
```

```python
print(a[::2])
```
