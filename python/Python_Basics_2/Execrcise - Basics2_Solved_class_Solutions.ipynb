## Exercise Question 2: Given a two list of equal size create a list of unique elements from both the lists into a seperate list

```python
lst1= [2, 3, 4, 5, 6, 7, 8]
lst2 = [4, 9, 16, 25, 36, 49, 64]
setA = set(lst1)
setB = set(lst2)
```

```python
result = list(setA.union(setB))
print("First List ",lst1)
print("Second List ",lst2)
print(result)

#using Union set operator
result = list(setA | setB)
print("First List ",lst1)
print("Second List ",lst2)
print(result)
```

```python
result = list(setA.intersection(setB))
print("First List ",lst1)
print("Second List ",lst2)
print(result)

#using Union set operator
result = list(setA & setB)
print("First List ",lst1)
print("Second List ",lst2)
print(result)
```

```python
result = list(setA.symmetric_difference(setB))
print("First List ",lst1)
print("Second List ",lst2)
print(result)

#using Union set operator
result = list(setA ^ setB)
print("First List ",lst1)
print("Second List ",lst2)
print(result)
```

```python
lst1= [2, 3, 4, 5, 6, 7, 8]
lst2 = [2,3]
setA = set(lst1)
setB = set(lst2)
result = setA.issubset(setB)
print("First List ",lst1)
print("Second List ",lst2)
print("SETA on SETA",result)

result = setB.issubset(setA)
print("First List ",lst1)
print("Second List ",lst2)
print("SETB on SETA",result)
```

#### Disjoint Sets


```python
#disjoint sets are sets that have no common elements
lst1= [2, 3, 4, 5, 6, 7, 8]
lst2 = [2,3]
setA = set(lst1)
setB = set(lst2)
result = setA.isdisjoint(setB)
print("First List ",lst1)
print("Second List ",lst2)
print("Disjoint SETA on SETA",result)

result = setB.isdisjoint(setA)
print("First List ",lst1)
print("Second List ",lst2)
print("Disjount SETB on SETA",result)
```
